Pure Python Implementation

```python
def edit_dist_py(s , t):
        m, n = len(s), len(t)
        # m x n DP chart
        dp = [[0]*(n + 1) for _ in range(m + 1)]
        # Base cases
        for i in range(m + 1):
            dp[i][0] = i 
        for j in range(n + 1):
            dp[0][j] = j
        for i in range(1, m + 1):
            for j in range(1, n + 1):
                if s[i - 1] == t[j - 1]:
                    substitution_cost = 0
                else:
                    substitution_cost = 1
                dp[i][j] = min(dp[i - 1][j] + 1,
                               dp[i][j - 1] + 1,
                               dp[i - 1][j - 1] + substitution_cost)
        return dp[m][n]
```

```python
from numba import njit
import numpy as np

@njit(nogil=True)
def edit_dist_nb_np(s , t):
        m, n = len(s), len(t)
        # m x n DP chart
        dp = np.zeros((m + 1, n + 1), dtype=np.int64)
        # Base cases
        for i in range(m + 1):
            dp[i][0] = i 
        for j in range(n + 1):
            dp[0][j] = j
        for i in range(1, m + 1):
            for j in range(1, n + 1):
                if s[i - 1] == t[j - 1]:
                    substitution_cost = 0
                else:
                    substitution_cost = 1
                dp[i][j] = min(dp[i - 1][j] + 1,
                               dp[i][j - 1] + 1,
                               dp[i - 1][j - 1] + substitution_cost)
        return dp[m][n]
```

```python
from numba.typed import List

@njit
def edit_dist_nb_tl(s , t):
        m, n = len(s), len(t)
        # m x n DP chart
        dp = List()
        for _ in range(m + 1):
            inner = List([0]*(n + 1))
            dp.append(inner)
        # Base cases
        for i in range(m + 1):
            dp[i][0] = i 
        for j in range(n + 1):
            dp[0][j] = j
        for i in range(1, m + 1):
            for j in range(1, n + 1):
                if s[i - 1] == t[j - 1]:
                    substitution_cost = 0
                else:
                    substitution_cost = 1
                dp[i][j] = min(dp[i - 1][j] + 1,
                               dp[i][j - 1] + 1,
                               dp[i - 1][j - 1] + substitution_cost)
        return dp[m][n]
```
